```python
def denoise(data):
    fft_data = np.fft.fft(data)
    abs_fft = np.abs(fft_data)
    sorted_index = np.argsort(-abs_fft)
    fft_data[sorted_index[100:]] = 0
    return np.fft.ifft(fft_data)
```

```python
denoise_data = denoise(data.values[:, 4])
plt.plot(denoise_data)
```
